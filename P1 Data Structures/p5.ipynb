```python
import hashlib
from time import gmtime, strftime

class Block:

    def __init__(self, timestamp, data, previous_hash):
      self.timestamp = timestamp
      self.data = data
      self.previous_hash = previous_hash
      self.hash = self.calc_hash(data)

    def calc_hash(self, data):
      sha = hashlib.sha256()

      hash_str = data.encode('utf-8')

      sha.update(hash_str)

      return sha.hexdigest()
    
    def __str__(self):
        return 'Block \nHash: {} \nTimestamp: {} \nData: {} \nPrevious Hash: {}'.format(self.hash, self.timestamp, self.data, self.previous_hash)

class Node:
   def __init__(self, data, previous_node) -> None:
      gmtTime = strftime("%H:%M %d/%m/%Y", gmtime())
      if previous_node is None:
        previous_hash = "0"
      else:
        previous_hash = previous_node.block.hash
      self.block = Block(gmtTime, data, previous_hash)
      self.previous = previous_node

      return

class chain:
  def __init__(self):
      self.head = None
      self.tail = None

  def append(self, data):
      if self.head is None:
          self.head = Node(data, None)
          self.tail = self.head
          return
      
      preivous_node = self.tail
      self.tail = Node(data, preivous_node)

  def PrintChain(self):
    # Prints from the last node/block to the first node/block
    while self.tail:
      print(self.tail.block)
      self.tail = self.tail.previous
      

    
```
